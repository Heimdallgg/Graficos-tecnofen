```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd
import numpy as np

# Inicializar la aplicación Dash
app = dash.Dash(__name__)

# Cargar el DataFrame
df = pd.read_csv('df_tcssr_completo.csv')  # Asegúrate de cambiar 'ruta_al_archivo.csv' por la ruta correcta

# Filtrar columnas para los menús desplegables, excluyendo 'Genotipo' y 'Estado Fenológico'
columnas_para_correlacion = df.columns.difference(['Genotipo', 'Estado Fenológico']).tolist()
genotipos_unicos = df['Genotipo'].unique()

# Diseño de la aplicación Dash
app.layout = html.Div([
    dcc.Dropdown(
        id='genotipo-dropdown',
        options=[{'label': genotipo, 'value': genotipo} for genotipo in genotipos_unicos],
        value=genotipos_unicos[0],
        clearable=False
    ),
    dcc.Dropdown(
        id='variable-1-dropdown',
        options=[{'label': col, 'value': col} for col in columnas_para_correlacion],
        value=columnas_para_correlacion[0],
        clearable=False
    ),
    dcc.Dropdown(
        id='variable-2-dropdown',
        options=[{'label': col, 'value': col} for col in columnas_para_correlacion],
        value=columnas_para_correlacion[1] if len(columnas_para_correlacion) > 1 else columnas_para_correlacion[0],
        clearable=False
    ),
    dcc.Graph(id='correlation-line-graph')
])

@app.callback(
    Output('correlation-line-graph', 'figure'),
    [Input('genotipo-dropdown', 'value'),
     Input('variable-1-dropdown', 'value'),
     Input('variable-2-dropdown', 'value')]
)
def update_graph(selected_genotipo, selected_var1, selected_var2):
    filtered_df = df[df['Genotipo'] == selected_genotipo]
    correlaciones = []

    # Calcular correlación para cada estado fenológico
    for estado in sorted(filtered_df['Estado Fenológico'].unique()):
        df_estado = filtered_df[filtered_df['Estado Fenológico'] == estado]
        correlacion = df_estado[[selected_var1, selected_var2]].corr().iloc[0, 1]
        correlaciones.append(correlacion)

    # Crear gráfico de líneas con la evolución de la correlación
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=sorted(filtered_df['Estado Fenológico'].unique()), y=correlaciones,
                             mode='lines+markers', name='Correlación'))
    fig.update_layout(title=f'Correlación entre {selected_var1} y {selected_var2} para {selected_genotipo}',
                      xaxis_title='Estado Fenológico',
                      yaxis_title='Valor de Correlación',
                      yaxis=dict(range=[-1, 1]))
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
