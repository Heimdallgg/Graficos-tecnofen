```python
# Asumiendo que df_tcsc_ps es tu DataFrame y ya está cargado correctamente
import numpy as np
# Filtramos el DataFrame para obtener solo los genotipos específicos
genotipos_interes = ['QUC 3635-2020', 'QUC 3492-2018', 'QUC 3526-2017']
df_filtrado = df_tcsc_ps[df_tcsc_ps['Genotipo'].isin(genotipos_interes)]

# Separamos las columnas numéricas y de texto
columnas_numericas = df_filtrado.select_dtypes(include=[np.number]).columns
columnas_texto = df_filtrado.select_dtypes(exclude=[np.number]).columns

# Calculamos los promedios solo para las columnas numéricas
promedios = df_filtrado.groupby(['Genotipo', 'Estado Fenológico'])[columnas_numericas].transform('mean')

# Preparamos las filas nuevas duplicando las filas originales
filas_nuevas = df_filtrado.copy()

# Asignamos los promedios calculados a las columnas numéricas en las filas nuevas
for col in columnas_numericas:
    filas_nuevas[col] = promedios[col]

# Para las columnas de texto, simplemente copiamos los valores (ya se copiaron al crear filas_nuevas)

# Concatenamos las filas nuevas al DataFrame filtrado
df_filtrado_expandido = pd.concat([df_filtrado, filas_nuevas]).sort_values(by=['Genotipo', 'Estado Fenológico'])

# Procedemos a unir este DataFrame expandido con el resto del DataFrame original
df_resto = df_tcsc_ps[~df_tcsc_ps['Genotipo'].isin(genotipos_interes)]
df_tcsc_ps = pd.concat([df_resto, df_filtrado_expandido]).sort_values(by=['Genotipo', 'Estado Fenológico'])

# Reset index
df_tcsc_ps.reset_index(drop=True, inplace=True)
```
